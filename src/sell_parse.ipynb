```python
#IMPORT NECESSARY LIBRARIES

import numpy as np
import pandas as pd
import os
import torch
```

```python
#Prepare sell data

#Turn text files into single dataframe w/o duplicates w.r.t store, date, and item
#Returns single matrix 'data' ordered by [store, date, time, quantity]
df17 = pd.read_csv('../input/famartsales/sale2017.txt', sep="\t")
df18 = pd.read_csv('../input/famartsales/sale2018.txt', sep="\t")
del df17['time']
del df18['time']
new_df17 = df17.groupby(['store','date','item'])['quantity'].sum().reset_index()
new_df18 = df18.groupby(['store','date','item'])['quantity'].sum().reset_index()
df = pd.concat([new_df17, new_df18], ignore_index = True)
info = df.values

#Extract store, date, product #'s for all stores combined
store_list = np.unique(info[:, 0])
date_list = np.unique(info[:, 1])
prod_list = np.unique(info[:, 2])

#Sort product and date list
prod_list = sorted(prod_list) 
date_list = sorted(date_list)

#Create dictionary where keys are the stores and values are each stores sell amount
stores, data = info[:, 0], info[:, 1:]
dict = {store: data[stores == store] for store in np.unique(stores)}

#Use counter and new_dict to make a dictionary where key: store, value: encoded products
new_dict = {}
count = 0
for x in store_list:
    sold = np.zeros((len(date_list), len(prod_list)), dtype=int)
    current_store = dict.get(x)
    for x in current_store:
        sold[date_list.index(x[0]), prod_list.index(x[1])] = x[2]
    new_dict[store_list[count]] = sold
    count = count + 1
```
