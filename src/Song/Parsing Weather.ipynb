```python
import numpy as np
import pandas as pd
import calendar
import pickle
```

```python
weth_dict = {}
for i in range(2017, 2019):
    for j in range(1, 13):
        days = calendar.monthrange(i,j)[1]
        file_name = str(i)+str(j).zfill(2)+'.csv'
        print(file_name)
        weth = pd.read_csv("../../data/Weather Data/"+file_name)
        weth = weth[["測站", "監測日期", "監測時間","溫度(℃)","相對溼度(%)","降水量(mm)"]].replace('-', np.nan).dropna().sort_values(by=["測站", "監測日期", "監測時間"])
        
        weth["溫度(℃)"] = weth["溫度(℃)"].astype('float64')
        weth["相對溼度(%)"] = weth["相對溼度(%)"].astype('float64')
        weth["降水量(mm)"] = weth["降水量(mm)"].astype('float64') > 0
        
        tw = weth.groupby(["測站", "監測日期"])["溫度(℃)", "相對溼度(%)"].mean()
        tw2 = weth.groupby(["測站", "監測日期"])["降水量(mm)"].max()
        
        location = np.unique(tw.index.get_level_values(0))
        
        for loc in location:
            loss_day = []
            for x in range(days):
                if str(i).zfill(2)+'/'+str(j).zfill(2)+'/'+str(x+1).zfill(2) not in tw.loc[loc].index:
                    loss_day.append(x+1)
            
            temp = np.concatenate((np.array(tw.loc[loc].values), np.reshape(tw2.loc[loc].values, (-1, 1))), axis=1)
            print("before:", loc, len(temp))
            print(loss_day)
            for day in loss_day:
                temp = np.insert(temp, day-1, np.array([np.nan]*3), 0)
            print("after:", loc, len(temp))
            if loc not in weth_dict:
                weth_dict[loc] = temp
            else:
                weth_dict[loc] = np.concatenate((weth_dict[loc], temp))
```

```python
pickle.dump(weth_dict, open("../../data/Weather Data/weather_parse.pkl", "wb"))
```
